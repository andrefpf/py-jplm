# Read a single PGX view

```python
from jplm import PGXHandler
import matplotlib.pyplot as plt 

reader = PGXHandler()
image = reader.read("../tests/pgx_files/bikes.pgx")

plt.imshow(image, cmap="grey")
plt.show()
```

# Read full PGX light field

```python
from jplm import LightField

lf = LightField.from_pgx("../../../.pgx/Bikes/")
```

```python
import matplotlib.pyplot as plt 
view = lf.get_view(0, 0, 0)
plt.imshow(view, cmap="grey")
plt.show()
```
